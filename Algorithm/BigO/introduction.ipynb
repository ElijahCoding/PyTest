```python
def sum1(n):
    final_sum = 0
    for x in range(n+1):
        final_sum += x
    return final_sum
```

```python
sum1(10)
```

```python
def sum2(n):
    return (n*(n+1))/2
```

```python
sum2(10)
```

```python
%timeit sum1(100)
```

```python
%timeit sum2(100)
```

```python
from math import log
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('bmh')

n = np.linspace(1, 10, 1000)
labels = [
    'Constant', 'Logarithm',
    'Liner', 'Log Linear',
    'Quardratic', 'Cubic',
    'Exponential'
]

big_o = [
    np.ones(n.shape), np.log(n),
    n, n*np.log(n), n**2, n**3,
    2**n
]

plt.figure(figsize=(12, 10))
plt.ylim(0, 50)

for i in range(len(big_o)):
    plt.plot(n, big_o[i], label = labels[i])
    
plt.legend(loc=0)
plt.ylabel('Relative Runtime')
plt.xlabel('n')
```

# O(1) Constant

```python
def func_constant(values):
    print(values[0])
```

```python
lst = [1,2,3,4,5,6]
func_constant(lst)
```

# O(n) Linear

```python
def func_lin(lst):
    for val in lst:
        print(val)
```

```python
func_lin(lst)
```

# O(n^2) Quadratic

```python
def func_quad(lst):
    for item_1 in lst:
        for item_2 in lst:
            print(item_1, item_2)
```

```python
func_quad(lst)
```

```python
def print_once(lst):
    for val in lst:
        print(val)
```

```python
print_once(lst)
```

```python
def print_2(lst):
    for val in lst:
        print(val)
    for val in lst:
        print(val)
```

```python
print_2(lst)
```

```python
def comp(lst):
    print(lst[0])
    
    midpoint = len(lst)/2
    for val in lst[:int(midpoint)]:
        print(val)
    
    for x in range(10):
        print('working')
```

```python
lst = [1,2,3,4,5]
comp(lst)
```

## O(1 + n/2 + 10) = O(n)

```python
def matcher(lst, match):
    for item in lst:
        if item == match:
            return True
    return False
```

```python
lst
```

```python
matcher(lst, 1) # O(1)
```

```python
matcher(lst, 100) # O(n)
```

```python
def create_list(n):
    new_list = []
    
    for num in range(n):
        new_list.append('new')
    return new_list
```
